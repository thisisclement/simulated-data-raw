```python
import os
import pandas as pd
from tqdm import tqdm
```

```python
DIR_INPUT = "./data/"

if not os.path.exists(DIR_INPUT):
    print(f"DIR_INPUT not present")

```

```python
len(os.listdir(DIR_INPUT))
```

```python
for i, file in tqdm(enumerate(os.listdir(DIR_INPUT))):
    if file not in ["customer_profiles.pkl", "terminal_profiles.pkl"]
    if i == 0:
        transaction_df = pd.read_pickle(os.path.join(DIR_INPUT, file))
    else:
        transaction_df = pd.concat([transaction_df, pd.read_pickle(os.path.join(DIR_INPUT, file))], ignore_index=True, sort=False)
```
